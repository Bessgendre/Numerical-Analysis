```python
# this is a program to calculate the DFT of a given signal, then compare DFT and FFT

import numpy as np
import matplotlib.pyplot as plt
import time

# import the package that can do matrix multiplication
from numpy import matmul as mm
```

```python
# write a function to calculate the DFT of a given singal with n points
def DFT(f):
    length = len(f)
    # initialize the DFT matrix
    DFT_matrix = np.zeros((length, length), dtype = complex)
    for i in range(length):
        for j in range(length):
            DFT_matrix[i][j] = np.exp(-2j * np.pi * i * j / length)
    # calculate the DFT of the given signal
    g = mm(DFT_matrix, f)
    return g

# define a function that do FFT on a given signal
def FFT(f):
    n = len(f)
    if n == 1:
        return f
    else:
        omega_n = np.exp(-2j * np.pi / n)
        omega = 1
        f_even = FFT(f[0::2])
        f_odd = FFT(f[1::2])
        g = np.zeros(n, dtype = complex)
        for i in range(int(n / 2)):
            g[i] = (f_even[i] + omega * f_odd[i]) / 2
            g[i + int(n / 2)] = (f_even[i] - omega * f_odd[i]) / 2
            omega = omega * omega_n
        return g
```

```python
# define the goal signal with 128 points
n = 128
x = np.linspace(0, 1, n)
f = 0.7 * np.sin(2 * np.pi * 2 * x) + np.sin(2 * np.pi * 5 * x)

# show the original signal
plt.figure(figsize = (10, 5))
plt.plot(x, f)
plt.title('Original Signal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# evaluate the time efficiency of DFT and FFT
DFT_start = time.time()
g_DFT = DFT(f)
DFT_end = time.time()
DFT_time = DFT_end - DFT_start
print("The time of DFT is: ", DFT_time, "s")

FFT_start = time.time()
g_FFT = FFT(f)
FFT_end = time.time()
FFT_time = FFT_end - FFT_start
print("The time of FFT is: ", FFT_time, "s")
```

```python
# plot the result of FFT
plt.figure(figsize = (10, 5))
plt.plot(x, g_FFT.real, label = 'real part')
plt.plot(x, g_FFT.imag, label = 'imaginary part')
plt.title('FFT')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.show()

# print the real part and imaginary part of the result FFT
for i in range(n):
    print("x_", i, " is: ", g_FFT.real[i], "\t y_", i, " is: ", g_FFT.imag[i])
```
