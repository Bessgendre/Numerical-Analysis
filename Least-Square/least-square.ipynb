```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def TrigonometricLeastSquare(data):
    b = data[:,1]
    A = np.array([[np.sin(x),np.cos(x)] for x in data[:,0]])
    coefficient = np.linalg.solve(np.dot(A.T,A),np.dot(A.T,b))
    return(coefficient)

def MeanSquareDiviation(coefficient, data):
    a,b = coefficient[0],coefficient[1]
    datapredict = np.array([a*np.sin(x) + b*np.cos(x) 
                       for x in data[:,0]])
    squarediviation = (datapredict - data[:,1])**2
    MSD = np.sum(squarediviation)/len(data)
    return MSD
```

```python
#测试
a, b = 2, 3

#创建数据点
datatest = [[x,a*np.sin(x)+ b*np.cos(x)] for x in
            np.linspace(0,np.pi,10)]
datatest = np.array(datatest)

#画图
plt.plot(datatest[:,0],datatest[:,1])

#计算，比较差异
coeff = TrigonometricLeastSquare(datatest)
MSD = MeanSquareDiviation(coeff, datatest)
print("a = ",coeff[0],", b = ",coeff[1],"\nMean Square Diviation = ",MSD)
```

```python
#作业题
dataX = np.arange(0.25,2.75,0.25)
dataY = np.array([1.284,1.648,2.117,2.718,3.427,2.798,
                  3.534,4.456,5.465,5.894])
data = np.column_stack((dataX,dataY))
plt.plot(data[:,0],data[:,1])

coeff = TrigonometricLeastSquare(data)
MSD = MeanSquareDiviation(coeff, data)
print("a = ",coeff[0],", b = ",coeff[1],"\nMean Square Diviation = ",MSD)
```
