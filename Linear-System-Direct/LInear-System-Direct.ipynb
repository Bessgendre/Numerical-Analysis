```python
import numpy as np

# 高斯列主元消去法
def GaussinColumnPivot(a, b):
    m, n = a.shape
    if ( m < n ):
        print("error, single solution not guaranteed")
    else:
        l = np.zeros((n,n))
        for i in range(n):
            if (a[i][i] == 0):
                print("error, solution does not exist")
                
        for k in range(n - 1):
            # 交换主元
            ChangePivot(a, b, k, n)
            
            for i in range(k + 1, n):
                l[i][k] = a[i][k] / a[k][k]
                for j in range(m):
                    a[i][j] = a[i][j] - l[i][k] * a[k][j]
                b[i] = b[i] - l[i][k] * b[k]
                
        # 开始回代
        x = np.zeros(n)
        x[n - 1] = b[n - 1] / a[n - 1][n - 1]

        for i in range(n - 2, -1, -1):
            for j in range(i + 1, n):
                b[i] -= a[i][j] * x[j]
            x[i] = b[i] / a[i][i]
            
        for i in range(n):
            print("x" + str(i + 1) + " = ", x[i])
        
def ChangePivot(a, b, k, n):           
    # 辅助函数，交换主元
    bad = 0
    for i in range(k, n):
        # 找出绝对值最大的
        if bad < np.fabs(a[i][k]):
            bad = a[i][k]
            maxn = i
    a[[k, maxn], :] = a[[maxn, k], :]
    b[k], b[maxn] = b[maxn], b[k]
```
